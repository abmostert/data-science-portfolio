```python
#encode the categories
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

X_ohe = ohe.fit_transform(moscow_data[['Apartment type', 'Metro station', 'Region', 'Renovation']])

cols_ohe = ohe.get_feature_names_out(['Apartment type', 'Metro station', 'Region', 'Renovation'])

encoded = pd.DataFrame(X_ohe, columns = cols_ohe, index=moscow_data.index)



moscow_data_encoded = pd.concat([moscow_data, encoded], axis = 1)


moscow_data_encoded.drop('Apartment type', axis=1, inplace=True)
moscow_data_encoded.drop('Metro station', axis=1, inplace=True)
moscow_data_encoded.drop('Region', axis=1, inplace=True)
moscow_data_encoded.drop('Renovation', axis=1, inplace=True)

moscow_data_encoded.head()

```

Given that i see nonlinear trends in the area and price, and that most variables are categories, i will be optimizing for log of price.

```python
moscow_data_encoded['log Price'] = np.log10(moscow_data_encoded['Price'])
moscow_data_encoded['log Area'] = np.log10(moscow_data_encoded['Area'])
moscow_data_encoded['log Kitchen area'] = np.log10(moscow_data_encoded['Kitchen area'])
moscow_data_encoded['log Living area'] = np.log10(moscow_data_encoded['Living area'])
moscow_data_encoded.drop('Price', axis=1, inplace=True)
moscow_data_encoded.drop('Area', axis=1, inplace=True)
moscow_data_encoded.drop('Kitchen area', axis=1, inplace=True)
moscow_data_encoded.drop('Living area', axis=1, inplace=True)

moscow_data_encoded.head()
```

```python
y_data = moscow_data_encoded['log Price']

x_data = moscow_data_encoded.drop('log Price', axis=1)

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x_data, y_data, test_size = 0.2)
```

```python
# create the model based upon ordinary least squares
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_training_data, y_training_data)
```

```python
from sklearn.utils import resample

n_bootstraps = 10
coef_boot = []
inter_boot = []

for i in range(n_bootstraps):
    x_boot, y_boot = resample(x_data, y_data, random_state=i)
    model_boot = LinearRegression().fit(x_boot, y_boot)
    coef_boot.append(model_boot.coef_)
    inter_boot.append(model_boot.intercept_)

coef_boot = np.array(coef_boot)
inter_boot = np.array(inter_boot)
```

```python
#Here the uncertainty is being calculated. Since the underlying distribution is not normal, we obtain the percentile
#ranges that are close to 2 sigma for a normal distribution. We also get the standard deviation for a comparison
coef_mean = coef_boot.mean(axis=0)
coef_mean_lower, coef_mean_upper = np.percentile(coef_boot, [2.5, 97.5], axis=0)
inter_mean = inter_boot.mean(axis=0)
inter_mean_lower, inter_mean_upper = np.percentile(inter_boot, [2.5, 97.5], axis=0)
coef_std  = coef_boot.std(axis=0, ddof=0)
inter_std  = inter_boot.std(axis=0, ddof=0)

for i in range(len(coef_mean)):

    if i == 0:
        print(f'The 1st coefficient ({moscow_data_encoded.columns[i]}) is {coef_mean[i]} +/- {coef_mean_upper[i]-coef_mean[i]}/{coef_mean[i]-coef_mean_lower[i]} (2.5, 97.5 percentile) & 2 times standard deviation {2*coef_std[i]}')
    elif i == 1:
        print(f'The 2nd coefficient ({moscow_data_encoded.columns[i]}) is {coef_mean[i]} +/- {coef_mean_upper[i]-coef_mean[i]}/{coef_mean[i]-coef_mean_lower[i]} (2.5, 97.5 percentile) & 2 times standard deviation {2*coef_std[i]}')
    elif i == 2:
        print(f'The 3rd coefficient ({moscow_data_encoded.columns[i]}) is {coef_mean[i]} +/- {coef_mean_upper[i]-coef_mean[i]}/{coef_mean[i]-coef_mean_lower[i]} (2.5, 97.5 percentile) & 2 times standard deviation {2*coef_std[i]}')
    else:
        print(f'The {i+1}th coefficient ({moscow_data_encoded.columns[i]}) is {coef_mean[i]} +/- {coef_mean_upper[i]-coef_mean[i]}/{coef_mean[i]-coef_mean_lower[i]} (2.5, 97.5 percentile) & 2 times standard deviation {2*coef_std[i]}')

print(f'The intercept is {inter_mean} +/- {inter_mean_upper-inter_mean}/{inter_mean-inter_mean_lower} (2.5, 97.5 percentile) & 2 times standard deviation {2*inter_std}')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

def metrics(y_true, y_pred, label=""):
    if y_true.name == "log_price":  # back-transform for price-scale metrics
        y_true_p = np.expm1(y_true)
        y_pred_p = np.expm1(y_pred)
        return pd.Series({
            "R2 (log-space)": r2_score(y_true, y_pred),
            "MAE (price)": mean_absolute_error(y_true_p, y_pred_p),
            "RMSE (price)": np.sqrt(mean_squared_error(y_true_p, y_pred_p))
        }, name=label)
    else:
        return pd.Series({
            "R2": r2_score(y_true, y_pred),
            "MAE": mean_absolute_error(y_true, y_pred),
            "RMSE": np.sqrt(mean_squared_error(y_true, y_pred))
        }, name=label)

results = metrics(y_valid.rename("log_price"), y_pred, "DecisionTree")
print(results)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(
    n_estimators=300,   # number of trees
    learning_rate=0.05, # smaller LR → needs more trees, usually better generalization
    max_depth=3,        # depth of each small tree (weak learner)
    subsample=1.0,      # set <1.0 (e.g. 0.8) for stochastic gradient boosting
    random_state=42
)

cv = cross_validate(
    gbr, x_data, y_data,
    cv=5,
    scoring={"r2": "r2", "rmse": "neg_root_mean_squared_error"},
    n_jobs=-1,
    return_estimator=True
)

r2_scores = cv["test_r2"]
rmse_scores = -cv["test_rmse"]  # flip sign

print("GBR 5-fold R²:   ", np.mean(r2_scores).round(3), "±", np.std(r2_scores).round(3))
print("GBR 5-fold RMSE: ", np.mean(rmse_scores).round(4), "±", np.std(rmse_scores).round(4))
```

```python
# Average feature importances across folds
importances = np.mean([est.feature_importances_ for est in cv["estimator"]], axis=0)
fi = (pd.DataFrame({"Feature": x_data.columns, "Importance": importances})
        .sort_values("Importance", ascending=False))
fi.head(20)
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import cross_validate
import numpy as np
import pandas as pd

hgb = HistGradientBoostingRegressor(
    learning_rate=0.08,      # a bit higher → fewer iterations needed
    max_leaf_nodes=31,       # small trees (fast, strong default-ish)
    max_iter=300,            # upper bound; early_stopping will stop sooner
    early_stopping=True,
    validation_fraction=0.1,
    random_state=42
)

cv = cross_validate(
    hgb, x_data, y_data,
    cv=3,  # faster than 5-fold and usually fine
    scoring={"r2":"r2","rmse":"neg_root_mean_squared_error"},
    n_jobs=-1,
    return_estimator=True
)

r2_mean  = np.mean(cv["test_r2"])
rmse_mean = -np.mean(cv["test_rmse"])
print(f"HGBR 3-fold R²: {r2_mean:.3f}  |  RMSE: {rmse_mean:.4f}")
```

```python
# pick the best fold by R^2
best_idx = int(np.argmax(cv["test_r2"]))
best_est = cv["estimator"][best_idx]

# 1) Exact params used (same as you set)
best_params = best_est.get_params()

# 2) Early-stopping result: how many boosting iterations actually used
# (only for HistGradientBoostingRegressor)
n_iter_used = best_est.n_iter_   # <-- this is often what you want

print("Best fold R²:", cv["test_r2"][best_idx])
print("Iterations used (early stopping):", n_iter_used)
print("Params:", best_params)
```
